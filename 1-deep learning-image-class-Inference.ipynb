```python
# Imports
from fastai.vision.all import *
from fastcore.all import *
```

```python
learn = load_learner('cdm_model.pkl')
```

```python
# learn.predict returns 
# 1. class string "cat", "dog", "forest", "monkey"
# 2. index - 0 for "cat", 1 for "dog", 2 for "forest", 3 for "monkey"
# 3. probablity - list of 4 tensors, match with index to get the probablity of the predicted class 
# dict(zip(categories, map(float, probs)))
def predict_image_category(image):
    category, index, probs = learn.predict(PILImage.create(image))
    # print("Tensor probs : ", probs)
    # percentage = "{:.2%}".format(a_number)
    fl_probs = ["{:.2%}".format(float(num)) for num in probs]
    # print ("Float probs: ", fl_probs)
    trained_for = ["cat", "dog", "forest", "monkey"]
    print(dict(zip(trained_for,  fl_probs)))
    if ((category in trained_for) and (probs[index] > 0.9900)) :
        str = f"Image of {category} with probablity {fl_probs[index]}"
        return str
    else :
        return "This is not an image of cat/dog/forest/monkey"
    
```

```python
dest = 'lion.jpg'
im = Image.open(dest)
im.to_thumb(256,256)
```

```python
predict_image_category(dest)
```
